task5: Prediction Using Decision Tree Algorithm

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Load the dataset (Replace 'your_dataset.csv' with the actual file path)
df = pd.read_csv("C:\\Users\\Madhusmita Panda\\Downloads\\Iris (1).csv")
df
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target labels

# Convert to DataFrame for better readability (optional)
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['target'])
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create the Decision Tree classifier and fit it to the training data
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)

#clf is a variable that holds the DecisionTreeClassifier model. clf means ="classifier"
```

```python
# Plot the decision tree
plt.figure(figsize=(10,8))
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```

```python
# Predict the test data
y_pred = clf.predict(X_test)

# Calculate the accuracy
accuracy = np.mean(y_pred == y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
