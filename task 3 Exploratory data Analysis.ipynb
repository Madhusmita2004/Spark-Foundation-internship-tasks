```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv(("C:\\Users\\Madhusmita Panda\\Downloads\\SampleSuperstore.csv"))
df
```

```python
#find the shape of the dataset
df.shape
```

```python
# Display the summary of the dataset
df.info()
```

```python
# Display the descriptive statistics
df.describe()
```

```python
# display missing value
missing_values=df.isnull().sum()
print("\nMissing values:")
missing_values
```

```python
#display duplicate value
df.duplicated().sum()
```

```python
#drop dulicate values
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
#category vs Sub Category
plt.figure(figsize=(10,6))
plt.bar('Sub-Category','Category', data=df)
plt.title('Category vs Sub Category')
plt.xlabel('Sub-Catgory')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyzing the distribution of categories in 'Category' column
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Category')
plt.title('Distribution of Categories')
plt.show()
```

```python
# Analyzing the distribution of sub-categories in 'Sub-Category' column
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Sub-Category', order=df['Sub-Category'].value_counts().index)
plt.title('Distribution of Sub-Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyzing the Sales by Category
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Category', y='Sales', estimator=sum)
plt.title('Total Sales by Category')
plt.show()
```

```python
# Analyzing the Profit by Category
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Category', y='Profit', estimator=sum)
plt.title('Total Profit by Category')
plt.show()
```

```python
# Analyzing the Sales by Region
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Region', y='Sales', estimator=sum)
plt.title('Total Sales by Region')
plt.show()
```

```python
# Analyzing the Sales by State
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='State', y='Sales', estimator=sum, order=df.groupby('State')['Sales'].sum().sort_values().index)
plt.title('Total Sales by State')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyzing the Profit by State
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='State', y='Profit', estimator=sum, order=df.groupby('State')['Profit'].sum().sort_values().index)
plt.title('Total Profit by State')
plt.xticks(rotation=90)
plt.show()
```

```python

# Scatter plot to analyze the relationship between Sales and Profit
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category')
plt.title('Sales vs Profit')
plt.show()
```

```python
# Identify meaningful patterns: Explore relationship
plt.figure(figsize=(4,2))
sns.pairplot(df)  
plt.show()
```

```python
#numerical analysis
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
print(numeric_columns)
```

```python
#display the histogram plot of numerical columns
df.hist(bins=50 ,figsize=(20,15))
plt.show();
```

```python
# Bivariate analysis
correlation_matrix = df[numeric_columns].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
